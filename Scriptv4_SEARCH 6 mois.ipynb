```python
import pandas as pd
from re import search, match
```

```python
def getv(url, m):
    reg = '.*\/' + m + '\/([a-z_A-Zéèèêàâäüùôöç0-9+ \'\.\-%,]*)\/.*'  
    ismatch = match(reg, url)
    if ismatch:
        value = ismatch.group(1)
    else:
        value = 'NaN'
    return value
def get_mdic(url):
    mdic = {}
    mlist = ['zoekterm','beginjaar','eindjaar','rubriek','eadid','findaidstatus','dao','lang',
                     'inventarisnr','level','scan-index','foto','page']
    for m in mlist:
        k = m
        v = getv(url, m)
        mdic[k] = v
    return mdic
```

```python
def parse(din):
    din.columns = ['visitorID','visitID','time','url']
    sl = []
    c = Counter(din)
    for i, r in din.iterrows():
        c.loop()
        rdic = {}
        reg = '.*\/zoekterm\/.*'
        if search(reg,r.url):
            rdic = get_mdic(r.url)
            rdic['visitID'] = r.visitID
            rdic['visitorID'] = r.visitorID
            rdic['time'] =  r.time
            rdic['url'] =  r.url
            sl.append(pd.Series(rdic))
    dout = pd.DataFrame(sl)
    return dout
```

```python
from re import sub
import re

def clean_text(text):
    if re.search("^[0-9\-\.\/]*$", text) != None : #on cherche à ne pas prendre les requêtes composées seulement de chiffres
        text = 'NULL'
    elif text != 'NaN':
        text = str(text)
        text = text.lower()
        text = sub('[^a-zéèèêàâäüùôöç0-9\']', ' ', text) #pour enlever les '+' ? et màj pour prendre en compte les accents*$
        text = str.replace(text,'é','e') #un peu laborieux mais fait le travail
        text = str.replace(text,'è','e')
        text = str.replace(text,'ê','e')
        text = str.replace(text,'à','a')
        text = str.replace(text,'â','a')
        text = str.replace(text,'û','u')
        text = str.replace(text,'ù','u')
        text = str.replace(text,'ô','o')
        text = str.replace(text,'ç','c')
        text = str.replace(text,'ö','o')
        text = str.replace(text,'ë','e')
        text = str.replace(text,'\'',' ')
        text = str.replace(text,'  ',' ')
        text = str.replace(text,'   ',' ')
        text = text.strip()
        #tlist = text.split(' ')
        #tlist.sort()
        #text = ' '.join(tlist)
    return text
```
